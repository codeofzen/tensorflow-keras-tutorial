```python
from keras.applications.resnet50 import ResNet50
from keras.preprocessing import image
from keras.applications.resnet50 import preprocess_input, decode_predictions
import numpy as np

from IPython.display import Image
```

```python
model = ResNet50(weights='imagenet')
```

```python
def predict(model, image_path):
    """Predict imagenet classes for the given image-url"""
    img = image.load_img(image_path, target_size=(224, 224))
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis=0)
    x = preprocess_input(x)

    return model.predict(x)


def print_predictions(prediction, top=3):
    decoded = decode_predictions(preds, top=top)[0]
    for p in decoded:
        print("%15s -  %2.2f %%" % (p[1], p[2]*100))
```

```python
image_url = './images/dog3.jpg'
Image(image_url)
```

```python
preds = predict(model, image_url)
print_predictions(preds, top=10)
```
